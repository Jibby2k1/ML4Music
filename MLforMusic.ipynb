#Generate Statistics

We want to generate some relevant statistics for music. We know that music shares time-based rules, therefore some "state" statistics will likely be useful to consider as features. We can start by generating sparse matrices to represent this data. Namely, we will generate Adjacency maps, and we know this will drastically reduce the size of the matrix because music tends to follow rules.

Instead of using weird set-based indeces, we can encode each combination to a unique ID. To achieve this, we can resort to number theory techniques that can help simplify our problem. We can relate each pitch combination to a unique combination of a 128-bit number (one-hot encoding of whether a note is active or not in a sequence).

This can become computationally complex, as there are (without consideration for musical rules), 128 C 10 notes that a human can play at any given time. This can evaluate to a massive number, but by using an adjacency map, we get to ignore all the ones we do not see.

# The Encoding Problem

The next problem to address, is how we make an adjacency matrix? When we recover values from the piano roll, we actually receive a "pitch index" which we can use to represent a note. Since the pitch indeces range from 1 to 128, we can encode any combination of co-existing pitches as a 128 bit number.

If this is hard to envision, imagine a smaller example:
Imagine a multiple choice exam where you can choose answers A, B, or C (not mutually exclusive, i.e. [A, B] or [B,C] ). You can easily count how many combinations there are: [], [A], [B], [C], [A, B], [A, C], [B, C], [A, B, C]. Which is 8 different combinations of the possible answers.

Let's get into set theory a bit, if we consider the problem of item ζ ⊂ {$ \mathcal{Q} $}, how many combinations of the elements are there? We can actually solve this by looking at the previous "multiple choice" example. The way to solve this problem is to analyze the following: {ζ} $\cup $ {$\mathcal{Q}-$ζ}.

Essentially, by considering each element's contribution to the set, we can evaluate the total number of combinations. Since each number is unique in this case, it will either be inside the set, or not inside the set, which are 2 states of being. Since there are 3 elements, that will correspond with $2^3$ total combinations.

Hopefully, it is easy to see how we can expand this to the case of notes existing in a segment.

```python
for en_index in range(len(encoding)-1):
  #print(encoding[en_index])
  if not(encoding[en_index] in state_map):
    state_map[encoding[en_index]] = {}
    state_map[encoding[en_index]][encoding[en_index+1]] =1;
  else:
    if not(encoding[en_index+1] in state_map[encoding[en_index]]):
      state_map[encoding[en_index]][encoding[en_index+1]] =1;
    else:
      state_map[encoding[en_index]][encoding[en_index+1]] +=1;
```

## Generating State Map Function

```python
#recall that note_seqs contains all the piano rolls, so we only need to pass in
#one piano roll at a time
def updateStateMap( st_map, song, window_size):

  #all the active pitches (non-zero velocity)
  active_pitches = []
  for pitches in piano_roll:
    active_pitches.append(np.where(np.array(pitches)!= 0)[0])

  #generate encoding for each combination of pitches currently being played in time
  encoding = []
  for pitches in active_pitches:
    encoding.append(tuple(pitches))

  #iterate through encoding list by index
  for en_index in range(len(encoding)-window_size):
    #previous states dictated by window size
    prevs = tuple(encoding[en_index::en_index+window_size])
    #if previous states is not a key in state map
    if not(prevs in st_map):
      st_map[prevs] = {}
      st_map[prevs][encoding[en_index+window_size+1]] =1;
    else:
      #if next state is not a key in previous state dict
      if not(encoding[en_index+window_size+1] in state_map[prevs]):
        st_map[encoding[prevs]][encoding[en_index+window_size+1]]  =1;
      #if both exist
      else:
        st_map[encoding[prevs]][encoding[en_index+window_size+1]] +=1;
```

### Cumulative State Map For Size 1 Window.

```python
#for each song's piano_roll, update the state map using window size of 1
for piano_roll in note_seqs:
  updateStateMap(state_map, piano_roll, 1)
```

```python
#for each song's piano_roll, update the state map using window size up to 10
max_window_length = 10
for ws in range(1, max_window_length+1):
  for piano_roll in note_seqs:
    updateStateMap(state_map, piano_roll, ws)
```


# Generating Histograms Problem

At this step, we would have the updated state maps. Now we want to visualize this information. Ideally, we can represent a next state using a distribution over the potential next values. If we can generate a histogram and treat that current state as a random variable (via the encoding) x, then we can find some distribution such that the true model ~ $\mathcal{D(x | \Theta)}$

Since x represents a state, we can use an encoding for x to generate the desired output, which we already have in the form of a tuple, however we would need to discretize this.

The main problem is that discretizing this would require a number of size $2^{128}$, and an extension of this problem to multiple previous states would require this number to be further appended in memory, that means that for $\mathcal{N}$ previous states, there would be $(2^{128})^{\mathcal{N}}$ bits necessary to represent this.

It is easy to see that a window size of 1 would require 2 longs to represent the number, and a window size of $\mathcal{N}$ would require $2 * \mathcal{N} $ longs to represent.

For plotting histograms, this is a big problem, as it will not be able to handle this amount of inputs or outputs.

## Markov Chains

While visualizing these histograms would be nice, and having access to them would allow us to summarize the data using much less information, they still provide us with a decent tool - the ability to make a prediction. In theory, we would have a distribution representing $ P(next | prev)$ which is proportional to $P(prev | next) * P(next)$ [Bayes Rule]. We can actually draw a similarity in the following way: $ P(\Theta_{next} | x_{prev}) ∝ P(x_{prev} | \Theta_{prev}) P(\Theta_{prev})$ meaning that if we have a model using the previous as an input, we can predict which next corresponds to this input.

While this is an amazing conclusion, this does not account for any of the states prior to the current previous state.

## Generating 'Local' State Maps - Motivation for features

We still have this count information available, perhaps it is possible to still use it in a meaningful way. Typically in music, notes do not just jump into a "far" octave out of the blue. Although it may be something to consider, it is not typical that this occurs in music.

Therefore it may be more appropriate to create "Local" State Maps, which map "Local" previous states to a "Local" next state. It should be evident that this will reduce the computation time, and even possibly the data size for a state.

$\mathcal{S} \to L_s\big[N(\mathcal{S}) \big] ⊂ N(\mathcal{S})$

Where $\mathcal{S}$ is the state, $N$ is the model that generates the next states, and $L_s$ is an operation that uses the local qualities (static time) of the state.

Local states in space also motivate the need for local states in time, which we can denote $L_t$.

Another name for the "local" states that we are describing, is simply a feature, which from now on we will call feature $L_s$ and $L_t$ as shorthand for "local state space" and "local state time" respectively.


### Examples in the context of Music Theory

1. A naive assumption is that we can choose $L_s$ such that the local states are represented by an octave. A better assumption would be to assume that local states represent a tonic.

2. A naive assumption is that a song will remain in the same key throughout its duration. A better assumption would be to segment the song into windows represented by their keys.
